```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 15})
```

```python
x = np.linspace(0,5,6)
y = 2*x + np.random.default_rng(12348).normal(0,1,len(x))
```

```python
z1 = np.polyfit(x,y,1)
p1 = np.poly1d(z1)
```

```python
z1 = np.polyfit(x,y,1,full=True)
z1
```

```python
z5 = np.polyfit(x,y,5)
p5 = np.poly1d(z5)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y,'o',label='data');

ax.legend()
ax.set(xlabel='x, input',ylabel='y, output')
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y,'o',label='data');

_x = np.linspace(-0.4,5.7,100)
ax.plot(_x,p1(_x),label='ax+b fit')

ax.legend()
ax.set(xlabel='x, input',ylabel='y, output')
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y,'o',label='data');

_x = np.linspace(-0.4,5.7,100)
ax.plot(_x,p1(_x),label='ax+b fit')
ax.plot(_x,p5(_x),label='polynomial (degree 5)')

ax.legend()
ax.set(xlabel='x, input',ylabel='y, output')
```
